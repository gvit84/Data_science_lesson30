Знаючи властивість рівномірного розподілу ми можемо знайти знайти щільність розподілу за наступною формулою:

f(x) = 1 / (b-a), де a і b - це межі інтервалу, на якому випадкова величина розподілена рівномірно.

Також математичне сподівання можна виразити через a і b:

М(Х) = (a+b) / 2

За умовою задачі М(Х) = 2, звідси:

(a+b) / 2 = 2  =>  a+b = 4

Також за умовою задачі ми знаємо, що:

P(X ≥ 3) = 0.4

Тому ймовірність того, що X менше або дорівнює 3, дорівнює:

P(X ≤ 3) = 1 - P(X ≥ 3) = 1 - 0.4 = 0.6

Оскільки X розподілена рівномірно, то:

P(X ≤ 3) = (3-a) / (b-a) = 0.6

Звідси отримуємо систему двох рівнянь з двома невідомими:

1. $a+b = 4$
2. $(3-a) / (b-a) = 0.6$

$a = 4 - b$

$(3 - (4 - b)) / (b - (4 - b)) = 0.6$

$(b - 1 ) / (2b - 4) = 0.6$

$b - 1 = 1.2b - 2.4$


$0.2b = 1.4$

**Отже:**

$b = 7$

$a = 4 - 7 = -3$

Отже, щільність розподілу випадкової величини Х буде:

**f(x) = 1 / (b-a) = 1 / (7-(-3)) = 1 / 10 = 0.1, якщо x належить інтервалу [-3,7]**

За межами цього інтервалу щільність дорівнюватиме нулю.
